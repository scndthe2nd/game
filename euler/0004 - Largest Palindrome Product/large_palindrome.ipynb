A palindromic number reads the same both ways. The largest palindrome made from the product of two $ 2 $-digit numbers is $ 9009 = 91 \times 99 $.
Find the largest palindrome made from the product of two $3$-digit numbers.


Separate the problem into parts;
- Find all products of 3 digit numbers
- Define Palendromes
- Check if the product is a palendrome
- Return largest. 

Count the number and split it in half(minus)


```python
def split_number(number):
    # Convert the number to a string
    number_str = str(number)
    
    # Calculate the length of the number string
    length = len(number_str)
    
    # Calculate the middle index
    middle_index = length // 2
    
    # Get the top half and bottom half
    top_half = number_str[:middle_index]
    
    # ignore the middle digit if length is odd 
    if length % 2 == 1:
        bottom_half = number_str[middle_index + 1:]
    else:
        bottom_half = number_str[middle_index:]
    return top_half, bottom_half

# Test the function with the given number
number = 1001001
top_half, bottom_half = split_number(number)

print(f"Top half: {top_half}")
print(f"Bottom half: {bottom_half}")
```

Reverse the bottom half

```python
def reverse_number(number):
    # Convert the number to a string
    number_str = str(number)
    
    # Reverse the string
    reversed_str = number_str[::-1]
    
    # Convert the reversed string back to an integer
    reversed_number = int(reversed_str)
    
    return reversed_number

# Test the function with the given number
number = 1234567
reversed_number = reverse_number(number)

print(f"Reversed number: {reversed_number}")
```

```python
def reverse_number(number):
    # Convert the number to a string
    number_str = str(number)
    
    # Reverse the string
    reversed_str = number_str[::-1]
    
    # Convert the reversed string back to an integer
    reversed_number = int(reversed_str)
    
    return reversed_number

def find_palindrome(number):
    number_str = str(number)
    return number_str == number_str[::-1]

palindromic_products = []

for valuea in range(99, 1000):
    for valueb in range(99, 1000):
        product = valuea * valueb
        if find_palindrome(product):
            palindromic_products.append((valuea, valueb, product))

# Sort the list by product
palindromic_products.sort(key=lambda x: x[2])

# Print sorted list
for factors in palindromic_products:
    print(f"{factors[0]} x {factors[1]} = {factors[2]}")

# Print the largest palindromic product
largest_palindrome = palindromic_products[-1]
print(f"The largest palindromic product of two three-digit numbers is {largest_palindrome[2]}, which is the product of {largest_palindrome[0]} and {largest_palindrome[1]}.")
```
